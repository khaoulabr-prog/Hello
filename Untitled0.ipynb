```python
"""
=============================================================
  MODÈLE DE SIMULATION ET OPTIMISATION — RENDEMENT FILTRATION
  Unité Filtration — OCP

  Formule de base :
    R = m(P2O5 rec) / m(P2O5) × 100
    m(P2O5 rec) = m(P2O5) - m(P2O5 perdu dans sérum gypse)

  m(P2O5 perdu) = m(gypse) × (seA,gypse) / 100

  Facteurs influençant (seA,gypse) identifiés par analyse :
    - %MS bouillie       → corrélation 0.493  (impact élevé)
    - %(CaO) phosphate   → corrélation 0.236  (impact moyen)
    - %(P2O5) phosphate  → corrélation 0.114  (impact faible)
    - m(gypse) produit   → corrélation 0.236  (impact moyen)
=============================================================
"""

import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy.optimize import minimize
import warnings
warnings.filterwarnings('ignore')


# ─────────────────────────────────────────────
#  1. MODÈLE EMPIRIQUE CALIBRÉ SUR DONNÉES RÉELLES
# ─────────────────────────────────────────────

# Coefficients de régression obtenus sur les 31 jours de données
# seA = f(%MS, %(P2O5), %(CaO), m(gypse))
COEF = {
    'intercept': -109.714,
    '%MS':        2.1793,
    '%(P2O5)':   -0.0125,
    '%(CaO)':    -2.1277,
    'm(gypse)':   0.0351,
}

# Plages opérationnelles observées (min/max données historiques)
RANGES = {
    '%MS':       (49.32, 51.11),
    '%(P2O5)':   (28.48, 29.58),
    '%(CaO)':    (47.89, 50.24),
    'm(gypse)':  (2880.0, 3086.0),
}

# Valeurs nominales (moyenne observée jours R >= 99%)
NOMINAL_OPTIMAL = {
    '%MS':       49.91,
    '%(P2O5)':   29.25,
    '%(CaO)':    48.84,
    'm(gypse)':  3004.3,
    'm(P2O5)':   586.0,
}

CIBLE_RENDEMENT = 99.0  # %


def predict_perte_serum(ms, p2o5, cao, m_gypse):
    """
    Prédit le % de P2O5 dans le sérum gypse à partir des paramètres opérationnels.
    Basé sur la régression linéaire multiple calibrée sur données OCP.
    """
    se = (COEF['intercept']
          + COEF['%MS']      * ms
          + COEF['%(P2O5)']  * p2o5
          + COEF['%(CaO)']   * cao
          + COEF['m(gypse)'] * m_gypse)
    return max(0.10, se)   # borne physique minimale


def calcul_rendement(ms, p2o5_pct, cao, m_gypse, m_p2o5_entrant):
    """
    Calcule le rendement complet à partir des paramètres de procédé.

    Entrées :
      ms              : % matière sèche bouillie (%)
      p2o5_pct        : % P2O5 dans la bouillie (%)
      cao             : % CaO dans le phosphate (%)
      m_gypse         : masse de gypse produit (kg)
      m_p2o5_entrant  : masse P2O5 entrant (kg)

    Sorties : dict avec toutes les variables intermédiaires
    """
    se_pct    = predict_perte_serum(ms, p2o5_pct, cao, m_gypse)
    m_perdu   = m_gypse * se_pct / 100
    m_rec     = m_p2o5_entrant - m_perdu
    rendement = (m_rec / m_p2o5_entrant) * 100

    return {
        '%MS':             ms,
        '%(P2O5)':         p2o5_pct,
        '%(CaO)':          cao,
        'm(gypse)':        m_gypse,
        'm(P2O5) entrant': m_p2o5_entrant,
        'se % gypse':      round(se_pct, 4),
        'm(P2O5) perdu':   round(m_perdu, 2),
        'm(P2O5) récupéré':round(m_rec, 2),
        'Rendement (%)':   round(rendement, 4),
        'Atteint cible':   rendement >= CIBLE_RENDEMENT,
    }


# ─────────────────────────────────────────────
#  2. ANALYSE DE SENSIBILITÉ
# ─────────────────────────────────────────────

def analyse_sensibilite(param_name, n_points=50):
    """
    Varie un paramètre sur sa plage opérationnelle (les autres fixés à nominal)
    et mesure l'impact sur le rendement.
    """
    lo, hi = RANGES[param_name]
    vals   = np.linspace(lo, hi, n_points)
    rends  = []

    for v in vals:
        p = dict(NOMINAL_OPTIMAL)
        p[param_name] = v
        res = calcul_rendement(
            p['%MS'], p['%(P2O5)'], p['%(CaO)'], p['m(gypse)'], p['m(P2O5)']
        )
        rends.append(res['Rendement (%)'])

    delta = max(rends) - min(rends)
    return vals, np.array(rends), delta


# ─────────────────────────────────────────────
#  3. OPTIMISATION — trouver les conditions R ≥ 99%
# ─────────────────────────────────────────────

def optimiser_conditions(m_p2o5_entrant=586.0):
    """
    Minimise la perte P2O5 (maximise le rendement) en respectant
    les contraintes opérationnelles.
    """
    x0 = [
        NOMINAL_OPTIMAL['%MS'],
        NOMINAL_OPTIMAL['%(P2O5)'],
        NOMINAL_OPTIMAL['%(CaO)'],
        NOMINAL_OPTIMAL['m(gypse)'],
    ]

    bounds = [
        RANGES['%MS'],
        RANGES['%(P2O5)'],
        RANGES['%(CaO)'],
        RANGES['m(gypse)'],
    ]

    def objectif(x):
        res = calcul_rendement(x[0], x[1], x[2], x[3], m_p2o5_entrant)
        return -res['Rendement (%)']   # maximiser → minimiser l'opposé

    # Contrainte : rendement >= 99%
    constraints = [{'type': 'ineq',
                    'fun': lambda x: calcul_rendement(x[0],x[1],x[2],x[3],m_p2o5_entrant)['Rendement (%)'] - 99.0}]

    result = minimize(objectif, x0, method='SLSQP', bounds=bounds, constraints=constraints,
                      options={'ftol': 1e-9, 'maxiter': 1000})

    optimal = calcul_rendement(result.x[0], result.x[1], result.x[2], result.x[3], m_p2o5_entrant)
    return result, optimal


# ─────────────────────────────────────────────
#  4. SIMULATION MONTE-CARLO — robustesse
# ─────────────────────────────────────────────

def simulation_monte_carlo(params_nominaux, bruit_pct=0.5, n=2000):
    """
    Simule la variabilité naturelle du procédé (bruit gaussien sur chaque paramètre)
    pour estimer la probabilité d'atteindre R >= 99%.
    """
    np.random.seed(42)
    rends = []
    for _ in range(n):
        noisy = {}
        for k in ['%MS', '%(P2O5)', '%(CaO)', 'm(gypse)']:
            sigma = (RANGES[k][1] - RANGES[k][0]) * bruit_pct / 100
            noisy[k] = np.clip(params_nominaux[k] + np.random.normal(0, sigma),
                               RANGES[k][0], RANGES[k][1])
        res = calcul_rendement(noisy['%MS'], noisy['%(P2O5)'], noisy['%(CaO)'],
                               noisy['m(gypse)'], params_nominaux['m(P2O5)'])
        rends.append(res['Rendement (%)'])
    rends = np.array(rends)
    return rends, (rends >= 99).mean() * 100


# ─────────────────────────────────────────────
#  5. RAPPORT GRAPHIQUE COMPLET
# ─────────────────────────────────────────────

def generer_rapport_graphique(output_path='rapport_optimisation.png'):
    fig = plt.figure(figsize=(18, 14), facecolor='white')
    gs  = gridspec.GridSpec(3, 4, figure=fig, hspace=0.5, wspace=0.4)

    params_plot = ['%MS', '%(P2O5)', '%(CaO)', 'm(gypse)']
    deltas = {}

    # --- Ligne 1 : sensibilité par paramètre ---
    for i, p in enumerate(params_plot):
        ax = fig.add_subplot(gs[0, i])
        vals, rends, delta = analyse_sensibilite(p)
        deltas[p] = delta
        nom_val = NOMINAL_OPTIMAL[p]

        couleurs = ['#1D9E75' if r >= 99 else '#E24B4A' for r in rends]
        ax.scatter(vals, rends, c=couleurs, s=20, zorder=3)
        ax.plot(vals, rends, color='#378ADD', lw=1.5, alpha=0.7)
        ax.axhline(99, color='#E24B4A', lw=1.5, ls='--', label='Cible 99%')
        ax.axvline(nom_val, color='#888780', lw=1, ls=':', alpha=0.8)
        ax.set_xlabel(p, fontsize=10)
        ax.set_ylabel('Rendement (%)', fontsize=9)
        ax.set_title(f'Sensibilité\nΔR = {delta:.3f}%', fontsize=10, fontweight='bold')
        ax.grid(alpha=0.3)
        if i == 0: ax.legend(fontsize=8)

    # --- Ligne 2 gauche : impact relatif ---
    ax_imp = fig.add_subplot(gs[1, :2])
    sorted_d = sorted(deltas.items(), key=lambda x: x[1], reverse=True)
    noms  = [x[0] for x in sorted_d]
    vals2 = [x[1] for x in sorted_d]
    bar_c = ['#E24B4A' if v > 0.3 else '#BA7517' if v > 0.15 else '#1D9E75' for v in vals2]
    bars  = ax_imp.barh(noms, vals2, color=bar_c, alpha=0.85)
    ax_imp.set_xlabel('ΔRendement max sur plage opérationnelle (%)', fontsize=10)
    ax_imp.set_title('Impact relatif des paramètres sur le rendement', fontsize=11, fontweight='bold')
    ax_imp.grid(axis='x', alpha=0.3)
    for bar, val in zip(bars, vals2):
        ax_imp.text(val + 0.002, bar.get_y() + bar.get_height()/2,
                    f'{val:.4f}%', va='center', fontsize=10)

    # --- Ligne 2 droite : Monte Carlo nominal ---
    mc_nom, prob_nom = simulation_monte_carlo(NOMINAL_OPTIMAL)
    result, optimal  = optimiser_conditions()
    mc_opt, prob_opt = simulation_monte_carlo({**NOMINAL_OPTIMAL,
                                               '%MS':      result.x[0],
                                               '%(P2O5)':  result.x[1],
                                               '%(CaO)':   result.x[2],
                                               'm(gypse)': result.x[3]})

    ax_mc = fig.add_subplot(gs[1, 2:])
    ax_mc.hist(mc_nom, bins=40, color='#378ADD', alpha=0.6, label=f'Conditions nominales\nP(R≥99%)={prob_nom:.1f}%')
    ax_mc.hist(mc_opt, bins=40, color='#1D9E75', alpha=0.6, label=f'Conditions optimales\nP(R≥99%)={prob_opt:.1f}%')
    ax_mc.axvline(99, color='#E24B4A', lw=2, ls='--', label='Cible 99%')
    ax_mc.set_xlabel('Rendement simulé (%)', fontsize=10)
    ax_mc.set_ylabel('Fréquence (n=2000)', fontsize=10)
    ax_mc.set_title('Simulation Monte-Carlo — robustesse du rendement', fontsize=11, fontweight='bold')
    ax_mc.legend(fontsize=9)
    ax_mc.grid(axis='y', alpha=0.3)

    # --- Ligne 3 : tableau de recommandations ---
    ax_tab = fig.add_subplot(gs[2, :])
    ax_tab.axis('off')

    headers = ['Paramètre', 'Valeur actuelle\n(moyenne)', 'Valeur optimale\npréconisée',
               'Plage opérationnelle', 'Impact sur R', 'Action']
    actuel  = [f"{NOMINAL_OPTIMAL['%MS']:.2f} %", f"{NOMINAL_OPTIMAL['%(P2O5)']:.2f} %",
               f"{NOMINAL_OPTIMAL['%(CaO)']:.2f} %", f"{NOMINAL_OPTIMAL['m(gypse)']:.0f} kg"]
    optim   = [f"{result.x[0]:.2f} %", f"{result.x[1]:.2f} %",
               f"{result.x[2]:.2f} %", f"{result.x[3]:.0f} kg"]
    plage   = [f"[{RANGES[p][0]} – {RANGES[p][1]}]" for p in params_plot]
    impacts = ['ÉLEVÉ ⬆', 'Faible', 'Moyen', 'Moyen']
    actions = ['Réduire %MS bouillie', 'Surveiller titrage', 'Optimiser ratio H₂SO₄/Pulpe', 'Contrôler vitesse filtre']

    rows = [[params_plot[i], actuel[i], optim[i], plage[i], impacts[i], actions[i]]
            for i in range(4)]

    table = ax_tab.table(
        cellText=rows, colLabels=headers,
        cellLoc='center', loc='center',
        bbox=[0, 0, 1, 1]
    )
    table.auto_set_font_size(False)
    table.set_fontsize(9)
    for (r, c), cell in table.get_celld().items():
        if r == 0:
            cell.set_facecolor('#378ADD')
            cell.set_text_props(color='white', fontweight='bold')
        elif impacts[r-1] == 'ÉLEVÉ ⬆' if r > 0 and r <= 4 else False:
            cell.set_facecolor('#FCEBEB')
        elif r % 2 == 0:
            cell.set_facecolor('#F1EFE8')

    ax_tab.set_title('Recommandations opérationnelles — paramètres à cibler',
                     fontsize=11, fontweight='bold', pad=10)

    fig.suptitle('Modèle d\'optimisation du rendement filtration P₂O₅ — OCP\n'
                 f'Rendement optimal calculé : {optimal["Rendement (%)"]:.4f} %',
                 fontsize=13, fontweight='bold', y=1.02)

    plt.savefig(output_path, dpi=150, bbox_inches='tight', facecolor='white')
    print(f"Rapport sauvegardé : {output_path}")
    return optimal, result


# ─────────────────────────────────────────────
#  6. EXÉCUTION PRINCIPALE
# ─────────────────────────────────────────────

if __name__ == '__main__':

    print("=" * 60)
    print("  SIMULATION RENDEMENT FILTRATION — OCP")
    print("=" * 60)

    # Conditions actuelles (exemple jour typique)
    print("\n--- Conditions actuelles (exemple) ---")
    res_actuel = calcul_rendement(
        ms=50.13, p2o5_pct=29.45, cao=49.50,
        m_gypse=3010.0, m_p2o5_entrant=589.0
    )
    for k, v in res_actuel.items():
        print(f"  {k:<25}: {v}")

    # Optimisation
    print("\n--- Optimisation des conditions opératoires ---")
    result_opt, cond_opt = optimiser_conditions(m_p2o5_entrant=589.0)
    print(f"\n  CONDITIONS OPTIMALES CALCULÉES :")
    print(f"  %MS           : {result_opt.x[0]:.3f} %  (actuel ~50.1%)")
    print(f"  %(P2O5)       : {result_opt.x[1]:.3f} %  (actuel ~29.4%)")
    print(f"  %(CaO)        : {result_opt.x[2]:.3f} %  (actuel ~49.5%)")
    print(f"  m(gypse)      : {result_opt.x[3]:.1f} kg (actuel ~3010 kg)")
    print(f"\n  Rendement optimal : {cond_opt['Rendement (%)']:.4f} %")
    print(f"  Perte sérum gypse : {cond_opt['se % gypse']:.4f} %")
    print(f"  Cible atteinte    : {'OUI ✓' if cond_opt['Atteint cible'] else 'NON ✗'}")

    # Monte Carlo
    print("\n--- Monte-Carlo (variabilité procédé) ---")
    rends_mc, prob = simulation_monte_carlo(NOMINAL_OPTIMAL)
    print(f"  Conditions nominales : P(R≥99%) = {prob:.1f}%")
    print(f"  Rendement moyen simulé : {rends_mc.mean():.3f}%")
    print(f"  Écart-type : {rends_mc.std():.4f}%")

    # Sensibilité
    print("\n--- Analyse de sensibilité ---")
    for p in ['%MS', '%(P2O5)', '%(CaO)', 'm(gypse)']:
        _, _, delta = analyse_sensibilite(p)
        print(f"  {p:<15}: ΔR = {delta:.4f} % sur la plage opérationnelle")

    # Rapport graphique
    print("\n--- Génération du rapport graphique ---")
    generer_rapport_graphique('/home/claude/rapport_optimisation.png')

    print("\n" + "=" * 60)
    print("  CONCLUSION")
    print("=" * 60)
    print("""
  Le paramètre le plus influent est le %MS de la bouillie.

  Causes principales des pertes P2O5 dans le sérum gypse :

  1. %MS bouillie trop élevé (>50%)
     → viscosité ↑ → mauvais lavage du gâteau → pertes ↑
     → ACTION : maintenir %MS entre 49.3% et 50.0%

  2. m(gypse) élevée → plus de surface de rétention d'acide
     → ACTION : optimiser ratio H2SO4/Pulpe (cadence réacteur)

  3. %(CaO) élevé → cristaux de gypse plus grossiers
     → Vérifier qualité phosphate entrant

  4. Lavage toile insuffisant (vide filtre, débit eau lavage)
     → Maintenir vide 300-350 mmHg, eau lavage ≥ 1.8× tonnage
""")
```
